# Deep leraning uygulaması için gerekli kütüphanlerin import edilmesi

```python
#Kullanılan Kütüphanelerin İmport edilmesi

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense, Flatten, BatchNormalization, Conv2D, MaxPool2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image_dataset_from_directory
from tensorflow.keras.preprocessing import image
from sklearn.metrics import confusion_matrix,classification_report



import seaborn as sns
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import roc_curve, auc, roc_auc_score
import itertools
import os
from tensorflow.keras.preprocessing import image
import numpy as np
import sys
import shutil
import random
import glob
import matplotlib.pyplot as plt
from PIL import Image
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
%matplotlib inline
```

```python
#Device and Hyperparamter ayarların verilmesi

# Device and parameters setting
device = len(tf.config.list_physical_devices('GPU'))

if device == 1:
    print(" Gpu " )
else:
    print(" Cpu " )

print("TensorFlow Version :", tf.__version__)
print("Keras Version      :", keras.__version__)


#Hyperparameters
image_size = 224
target_size = (image_size, image_size)
input_shape = (image_size, image_size, 3)
batch_size = 32
epochs=20

```

```python
#Dataset Path gösterimi ve train folder ,test_folder,valid_folder olşturulması
dataset_root= 'D:\Dataset'

train_folder =os.path.join(dataset_root,'train')
valid_folder =os.path.join(dataset_root,'valid')
test_folder =os.path.join(dataset_root,'test')
```

```python
train_data = tf.keras.utils.image_dataset_from_directory(
                                                         train_folder,
                                                         
                                                         seed=123,
                                                         image_size=(224,224),
                                                         batch_size=batch_size)
```

```python
#Agumanetin başlamadan önce İlk gösterim
img = Image.open(os.path.join(train_folder,'Mug','Mug (185).jpg'))
print(np.asarray(img, dtype=np.uint8).shape)
plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
#Agumentin için custom butun imajlara uygulanacak ayarlar
#aşağıdaki # sonraki acıklamaları Tensorflow sitesinde custom verilebilecek değerler olarak buldum bu değerleri biraz daha 
#yükseltim cünkü trainin aşamasında model accoury %100 olarak bitebiliryor yani overfit oluyor olarak düşünüyorum
#bu neden ile overfit biraz önlemek amaclı değerleri yükseltim
datagen =ImageDataGenerator(
                             rotation_range=35, # rotation10
                              width_shift_range=0.5, # horizontal shift0.2
                              height_shift_range=0.5, # vertical shift0.2
                              zoom_range=0.5, # zoom0.2
                              rescale=1./255,
                              horizontal_flip=True, # horizontal flip
                              brightness_range=[0.2,1.2]) # brightness0.2,1.2
```

```python
#Agumantin oluşturulan custum datagen değişkeni ile train , valid, test dataloader lar oluşuturuluyor
#900 mug ve 900 adet nomug train seti için yukarıda verildi,
#300 er adet  mug ve nomug verildi
#100 er adet test için verildi 
#normalde %800 train, %10 valid ve % 10 test klasıorlerine ayrılır fakat deep learning modelleri
#için normal şartlarda 2000k civarlarında datasetleri ile çelşıyor image az olduğu için bu şekilde birazda 
#valid klasorune daha fazla verdiö
train_aug = datagen.flow_from_directory(train_folder,
                                         target_size=(image_size, image_size),
                                         batch_size=batch_size,
                                         shuffle = True,
                                         classes=['mug', 'NoMug'])


valid_aug = datagen.flow_from_directory(valid_folder,
                                         target_size=(image_size, image_size),
                                         batch_size=batch_size,
                                         shuffle = True,
                                         classes=['mug', 'NoMug'] )


test_aug = datagen.flow_from_directory(test_folder,
                                         target_size=(image_size, image_size),
                                         batch_size=batch_size,
                                         shuffle = False,
                                         classes=['mug', 'Nomug'] )
```

```python
#Agumentin aşamasından sonra Örnek  Gösterim
for i in range(9):
    #plt.subplot(330 + 1 + i)
    ax = plt.subplot(3, 3, i + 1)
    images, labels = train_aug.next()
    image = (images[0]*255).astype('uint8')
    label = (labels[0])
    plt.imshow(image)
    plt.axis("off")
```

```python
#Resimlerin ve etiketlerin boyut bilgisi
for image_batch, labels_batch in train_aug:
  print(image_batch.shape)
  print(labels_batch.shape)
  break
```

```python
#Imageların bulunan classların yazırılması
class_names = train_data.class_names
print(class_names)
```

```python
#Model Oluşturma

model = Sequential([
    Conv2D(filters=32, kernel_size=(3, 3), activation='relu', 
           padding = 'same',input_shape=(224,224,3)),
    MaxPool2D(pool_size=(2, 2), strides=2),

    Conv2D(filters=64, kernel_size=(3, 3), activation='relu', padding = 'same'),
    MaxPool2D(pool_size=(2, 2), strides=2),

    Conv2D(filters=128, kernel_size=(3, 3), activation='relu', padding = 'same'),
    MaxPool2D(pool_size=(2, 2), strides=2),

    Conv2D(filters=256, kernel_size=(3, 3), activation='relu', padding = 'same'),
    MaxPool2D(pool_size=(2, 2), strides=2),

    Conv2D(filters=512, kernel_size=(3, 3), activation='relu', padding = 'same'),
    MaxPool2D(pool_size=(2, 2), strides=2),

    Conv2D(filters=1024, kernel_size=(3, 3), activation='relu', padding = 'same'),
    MaxPool2D(pool_size=(2, 2), strides=2),


    Flatten(),
    Dense(128, activation='relu'),
    Dense(units=2, activation='softmax'),
    
])
```

```python
model.summary()
```

```python
#Optimazer ayarlarının modele verilmesi
model.compile(optimizer=Adam(learning_rate=0.000001), loss='categorical_crossentropy', metrics="categorical_accuracy")
```

```python
#callback tensor modelinin training aşamasında overfitting önlemek amaclı tavsiyesidir.
#https://www.tensorflow.org/api_docs/python/tf/keras/callbacks/EarlyStopping

callback = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=3)
model.fit(x=train_aug,
          steps_per_epoch=len(train_aug), 
          validation_data=valid_aug,
          validation_steps=len(valid_aug),
          epochs=epochs, 
          verbose=2,
           callbacks=[callback])
```

```python
results = model.evaluate(test_aug, verbose=1)
print("Test Accuracy: {:.2f}%".format(results[1] * 100))
```

```python
#Örnek model kaydeliyor
model.save('mug.h5')
```

```python
#Roc curve için işlemler
classes = dict((v, k) for k, v in test_aug.class_indices.items())
num_classes = len(classes)
num_samples = len(valid_aug)


input_size = 224
y_true = np.empty([num_samples, 2])
y_pred = np.empty([num_samples, 2])
X = np.empty([input_size, input_size, 3])

# test seti için bir örnek üretilim   
for i in range(num_samples):
    prog = ' Progress: '+str(i+1)+'/'+str(num_samples)
    X, y_true = next(valid_aug)
    y_pred = model.predict(X, steps=1)
    sys.stdout.write('\r'+prog)
sys.stdout.write('\rDone                ')
```

```python
from sklearn.metrics import roc_curve, auc
fpr = dict()
tpr = dict()
thresholds = dict()
roc_auc = dict()

# her bir class için nnegatif /pozitif hesaplanması
for i in range(num_classes):
    fpr[i], tpr[i], thresholds[i] = roc_curve(y_true[:, i], y_pred[:, i], drop_intermediate=False) 
    #fpr ve tpr ve treshold orneklemden oluştumak
    roc_auc[i] = auc(fpr[i], tpr[i])
    print(auc(fpr[i], tpr[i]))
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20, 10))
for i in range(num_classes):
    ax[i].plot(fpr[i], tpr[i], color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc[i])
    ax[i].plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    ax[i].set_xlim([0.0, 1.0])
    ax[i].set_ylim([0.0, 1.03])
    ax[i].set_xlabel('False Positive Rate', fontsize=20)
    ax[i].set_ylabel('True Positive Rate', fontsize=20)
    ax[i].set_title('ROC Curve for class: '+classes[i], fontsize=20)
    ax[i].legend(loc="lower right", prop={'size': 20})
```

```python
def predict_image(filename,model):
    img_ = image.load_img(filename, target_size=(224, 224))
    img_array = image.img_to_array(img_)
    img_processed = np.expand_dims(img_array, axis=0) 
    img_processed /= 255.   
    
    prediction = model.predict(img_processed)
    index = np.argmax(prediction)
    klass=class_names[index]
    probability=prediction[0][index]*100
    plt.title("Prediction - {}".format(class_names[index]))
    plt.imshow(img_array)

    print(f'Bu Görüntü tahmini {klass} olma olasılığı {probability:6.2f} %')
```
